## 1. Introduction

This analysis is for case study 1 from the Google Data Analytics Certificate (Cyclistic). We will be using the Divvy data sets for the case study. The purpose of this script is to consolidate downloaded Divvy data into a single data frame and then conduct simple analysis to help answer the key question: “In what ways do subscriber and customer riders use Divvy bikes differently?”

### 1.1 Import Libraries

The following libraries will be used for the analysis.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
%matplotlib inline
```

### 1.2 Load and inspect datasets

```python
# load Q1 data
q1_2019 = pd.read_csv("Divvy_Trips_2019_Q1.csv")
print("Shape of q1_2019:", q1_2019.shape)
q1_2019.head()
```

```python
# load Q2 data
q2_2019 = pd.read_csv("Divvy_Trips_2019_Q2.csv")
print("Shape of q2_2019:", q2_2019.shape)
q2_2019.head()
```

```python
# load Q3 data
q3_2019 = pd.read_csv("Divvy_Trips_2019_Q3.csv")
print("Shape of q3_2019:", q3_2019.shape)
q3_2019.head()
```

```python
# load Q4 data
q4_2019 = pd.read_csv("Divvy_Trips_2019_Q4.csv")
print("Shape of q4_2019:", q4_2019.shape)
q4_2019.head()
```

### 1.3 Data cleaning

#### 1.3.2 Rename col names of q2_2019 dataframe to be consistent with colnames of other dfs

```python
q2_2019.rename(columns = {'Rental_Details_Rental_ID':'trip_id', 'Rental_Details_Local Start_Time':'start_time',
                          'Rental_Details_Local_End_Time':'end_time', 'Rental_Details_Bike_ID':'bikeid',
                          'Rental_Details_Duration_In Seconds_Uncapped':'tripduration',
                          'Rental_Start_Station_ID':'from_station_id', 'Rental_Start_ Station_Name':'from_station_name',
                          'Rental_End_Station_ID':'to_station_id', 'Rental_End_Station_Name':'to_station_name',
                          'User_Type':'usertype', 'Member_Gender':'gender',
                          'Member_Details_Member_Birthday_Year':'birthyear'}, inplace=True)
```

```python
q2_2019.columns
```

#### 1.3.3 Inspect data frames to look for inconsistencies

```python
q1_2019.dtypes
```

```python
q2_2019.dtypes
```

```python
q3_2019.dtypes
```

```python
q4_2019.dtypes
```

#### 1.3.4 Convert trip_id, start_time, and end_time columns data types to object to stack correctly in final dataframe

```python
q1_2019["trip_id"] = q1_2019["trip_id"].astype(str)
q1_2019["start_time"] = q1_2019["start_time"].astype("datetime64")
q1_2019["end_time"] = q1_2019["end_time"].astype("datetime64")

q1_2019.dtypes
```

```python
q2_2019["trip_id"] = q2_2019["trip_id"].astype(str)
q2_2019["start_time"] = q2_2019["start_time"].astype("datetime64")
q2_2019["end_time"] = q2_2019["end_time"].astype("datetime64")

q2_2019.dtypes
```

```python
q3_2019["trip_id"] = q3_2019["trip_id"].astype(str)
q3_2019["start_time"] = q3_2019["start_time"].astype("datetime64")
q3_2019["end_time"] = q3_2019["end_time"].astype("datetime64")

q3_2019.dtypes
```

```python
q4_2019["trip_id"] = q4_2019["trip_id"].astype(str)
q4_2019["start_time"] = q4_2019["start_time"].astype("datetime64")
q4_2019["end_time"] = q4_2019["end_time"].astype("datetime64")

q4_2019.dtypes
```

### 1.4 Merge all dataframes into single dataframe

```python
all_trips = pd.concat([q1_2019, q2_2019, q3_2019, q4_2019])

print("Shape of all_trips:", all_trips.shape)
all_trips.head()
```

### 1.5 Inspect new dataframe

```python
all_trips.columns
```

```python
all_trips.dtypes
```

```python
all_trips.describe(include=(object))
```

```python
# Drop 'tripduration' col and calculate it from start and end time
drop_cols = ["tripduration"]
all_trips.drop(columns=drop_cols, inplace=True)
# Create Cols for Year, Month, Day, and Day_name from the Data Frame
all_trips["year"] = all_trips["start_time"].dt.year
all_trips["month"] = all_trips["start_time"].dt.month
all_trips["day"] = all_trips["start_time"].dt.day
all_trips["day_of_week"] = all_trips["start_time"].dt.day_name()

all_trips.head()
```

<div id='eda'></div>

## 2. Data Analysis

### 2.1 Compute tripdurarion

```python
# Compute trip duration and cast as datetime object
all_trips["tripduration"] = (all_trips["end_time"] - all_trips["start_time"]).astype('timedelta64[m]')
```

### 2.2 Descriptive analysis on tripduration

```python
# Compute summary statistics of trip duration
all_trips["tripduration"].describe()
```

### 2.3 Remove trips with negative tripduration

```python
# Remove trips with negative duration which represent period when bike are on repair
tripduration = all_trips['tripduration']
all_trips_2 = all_trips.query('tripduration >= 0')
all_trips_2.dropna() # Drop missing values
```

### 2.4 Conduct descriptive analysis

#### 2.4.1 descriptive statistics for tripduration

```python
# summary statistics of trip duration
all_trips_2["tripduration"].describe()
```

#### 2.4.2 summary statistics of tripduration by usertype

```python
# compute summary statistics by user type of trip duration
all_trips_2.groupby(all_trips_2["usertype"]).tripduration.agg(["mean", "median", "max", "min"])
```

#### 2.4.3 average tripduration by usertype and day of the week

```python
# compute average trip duration by user type and by the day of the week
result_1 = all_trips_2.groupby(["usertype", "day_of_week"])["tripduration"].aggregate("mean").reset_index()
result_1
```

#### 2.4.4 number of trips by usertype and day of the week

```python
# compute average number of rides by user type and by the day of the week
result_2 = all_trips_2.groupby(["usertype", "day_of_week"])["trip_id"].aggregate("count").reset_index()
result_2
```

<div id='viz'></div>

## 3. Visualization

```python
# Create bar chart of `usertype`
sns.barplot(
    x='day_of_week',
    y='trip_id',
    hue='usertype',
    data=result_2,
)
plt.xlabel("Day of the week")
plt.ylabel("Number of rides")
plt.title("Number of Rides by Usertype by Week of the Day");

# save image 👇
plt.savefig("num_ride.jpg", dpi=150)
```

```python
sns.barplot(
    x='day_of_week',
    y='tripduration',
    hue='usertype',
    data=result_1,
)
plt.xlabel("Day of the week")
plt.ylabel("Average trip rides")
plt.title("Average Trip Duration by Usertype by Week of the Day");
plt.savefig("Average_tripduration.jpg", dpi=150)
```

## 4. Conclusions/Recommendations

Based on the analysis and findings above, maximizing the number of subscribers will go a long way in ensuring future success of the company. Below are some key insights derived;

    Subscribers recorded the highest numbers of rides for all the weekdays.
    Customers have higher number of trip duration across all weekdays.

Therefore, future marketing campaigns should focus on converting customers to subscribers for the success of the company.
